```python
import pandas as pd
```

```python
adidas=pd.read_csv('adidas.csv')
```

```python
adidas
```

```python
adidas.columns
adidas=adidas.drop(columns=['average_rating', 'reviews_count', 'crawled_at','selling_price', 'original_price'])
adidas.head(1)
```

```python
target=.98

columnas_multi_nivel=[]
columnas_un_nivel=[]


for c_ in adidas.columns:
    x=adidas.loc[: , [c_]].drop_duplicates()
    valores_unicos=x.shape[0]
    
    if (valores_unicos==1):
        columnas_un_nivel.append(c_)
    elif(valores_unicos/adidas.shape[0]) >= target:
        columnas_multi_nivel.append(c_)
   
```

```python
columnas_multi_nivel
```

```python
"""Primer inciso"""
adidas=adidas.drop(columns=columnas_multi_nivel)
```

```python
 adidas
```

```python
"""INCISO 2"""
```

```python
columnas_mid_valores=[]

for c_ in adidas.columns:
    x=adidas.loc[: , [c_]].drop_duplicates()
    valores_unicos=x.shape[0]
    if((valores_unicos>=3) & (valores_unicos<=6)):
        columnas_mid_valores.append(c_)
```

```python
columnas_mid_valores
```

```python
dfcat=adidas.loc[:,['category','name']].groupby(['category'], as_index=False).count()
```

```python
dfcat
```

```python
dfcat.iloc[0,0]
```

```python
adidasMod=adidas.replace(dfcat.iloc[0,0],"Others")
```
