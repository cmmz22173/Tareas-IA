```python
""""
    AQUI SE UTILIZO UN CSV
    
    Son las notas finales de un curso X"""
import pandas as pd
df=pd.read_csv("Notas.csv")
df
```

```python
"""Se necesita conocer la media de calificacion para conocer el rendimiento de cada la seccion en general"""

print("Promedio")
df['final grade'].mean()
```

```python
"""Se desea conocer los cinco estudiantes con mejor calificacion del curso para darles una beca estudiantil"""

df.sort_values(by=['final grade'],ascending=False).head()
```

```python
""""Se desea conocer el porcentaje de estudiantes por genero en el curso"""

df1 = (df.groupby(['gender'])['gender'].count()/df.id.count())*100
df1
```

```python
""""Se desea saber el porcentaje de estudiantes aprobados"""

aprobados=(df['final grade'][df['final grade']>=65].count()/df.id.count())*100
reprobados=(df['final grade'][df['final grade']<65].count()/df.id.count())*100

tabRend={
    'aprobados':[aprobados],
    'reprobados': [reprobados]
}

tr=pd.DataFrame(tabRend)
print('Porcentaje % de aprobados/reprobados')
tr
```

```python
"""
   AQUI SE UTILIZO UN JSON
   
   El contexto de esta informacion que quiero trabajar es sobre los graduados de un periodo X en un pais Y, y supongamos que las 
   universidades son de un solo pais con grandes cantidades de extranjeros, porque mockaroo no tenia algo tan especifico 
   entonces las que salen son globales. Ademas el join no es consistente porque los dos archivos se generaron aleatorio y de 
   1000 univeridades que tiene el json vs 200 graduados solo hicieron match 19. Quitando el Dataframe de Universidades.Json 
   no habra tanto problema(si lo quiere probar para que funcione mejor)"""
  
"""dfJson=pd.read_json("Graduados.json")"""

dfJson1=pd.read_json("Graduados.json")
dfJson2=pd.read_json("Universidades.json")

dfJson = pd.merge(dfJson1, dfJson2, how='inner', on='college' )
dfJson
```

```python
"""Se desea conocer la el promedio de edad con la que se graduo esta promocion del periodo X"""

print("Edad promedio de graduacion")
dfJson['age'].mean()
```

```python
"""Se desea saber el top 5 nacionalidades con mejor promedio academico de graduacion de este periodo X"""

dfJson[['nationality','graduation rate']].groupby(['nationality'])['graduation rate'].mean().sort_values(ascending=False).head(5)
```

```python
"""Se desean conocer las 3 universidades con mas graduados este periodo X 
   --Nota: Dado que solo hay 200 y demasiadas universidades puede que no cuenten con mas de 2 graduados por universidad
     por eso considero que con datos reales la pregunta tendria mas sentido  """

dfJson.groupby(['college'])['college'].count().sort_values(ascending=False).head(3)
```
